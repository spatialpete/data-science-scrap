```python
#https://stackoverflow.com/questions/28017091/will-pandas-dataframe-object-work-with-sklearn-kmeans-clustering
from sklearn import cluster
import sklearn
#X_iris, y_iris = datasets.load_iris(return_X_y=True)
#k_means = cluster.KMeans(n_clusters=3)
#k_means.fit(X_iris)


#dropping non-numeric columns
dfkmeans = dfnozero[['Description_Cat','WoCost','WoEquipCost','WoLaborCost','WoMatCost']]

# Convert DataFrame to matrix
mat = dfkmeans.values
# Using sklearn
km = sklearn.cluster.KMeans(n_clusters=5)
km.fit(mat)
# Get cluster assignment labels
labels = km.labels_
# Format results as a DataFrame
results = pd.DataFrame([dfkmeans.index,labels]).T
#results = pd.DataFrame(data=labels, columns=['cluster'], index=collapsed.index)
results
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(dfkmeans)
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(dfkmeans['WoEquipCost'], dfkmeans['WoLaborCost'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
# Cutting outliers
# https://www.kite.com/python/answers/how-to-remove-outliers-from-a-pandas-dataframe-in-python
from scipy import stats
z_scores = stats.zscore(dfkmeans)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_df = dfkmeans[filtered_entries]
```

```python
new_df
```

```python
# plotting data without outliers
kmeans = KMeans(n_clusters=3).fit(new_df)
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(new_df['WoEquipCost'], new_df['WoLaborCost'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```
